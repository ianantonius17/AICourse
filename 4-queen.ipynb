```python
r = c = 4

def printBoard(board):
    for row in board:
        for col in row:
            print (" |",col,end="")
        print(" |")

def printSteps(boardArray):
    for b in boardArray :
        printBoard(b)

def checkRow(board , row):
    for i in range(c):
        if board[row][i] == 1:
            return False
    return True
def checkCol(board , col):
    for i in range(r):
        if board[i][col] == 1:
            return False
    return True    
    

def checkDiagonal(board,row,col):
    for i in range(r):
        if i == 0:
            continue
        if row-i >= 0 and col-i >= 0:
            if board[row-i][col-i] == 1:
                return False
        if row+i < 4 and col+i < 4:
            if board[row+i][col+i] == 1:
                return False
        if row-i >= 0 and col+i < 4:
            if board[row-i][col+i] == 1:
                return False
        if row+i < 4 and col-i >= 0:
            if board[row+i][col-i] ==1:
                return False
    
    return True
        
def checkPath(board,row,col):
    return checkRow(board,row) and checkCol(board,col) and checkDiagonal(board,row,col)


def placeQueen(board,row):
    
    if row >= r:
        return True
    printBoard(board)
    print()
    for i in range(c):
        
        if(checkPath(board,row,i)==True):
            board[row][i] = 1
            
            if placeQueen(board,row+1):
                return True
            board[row][i] = 0
    
    return False
    
```
